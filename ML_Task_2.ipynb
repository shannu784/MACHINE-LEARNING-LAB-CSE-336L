```python
# 4) Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
import matplotlib.pyplot as plt

x = []
y = []
for line in open('sample.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(lines[0])
    y.append(int(lines[1]))

plt.title("Students Marks")
plt.xlabel('Name')
plt.ylabel('Marks')
plt.yticks(y)
plt.plot(x, y, marker = 'o', c = 'g')

plt.show()
```

```python
# 5) Write a Python program to plot two or more lines with legends, different widths and colors.

x1 = [1, 2, 3, 4, 5]
y1 = [3, 5, 7, 2, 1]
x2 = [1, 2, 3, 4, 5]
y2 = [5, 7, 2, 3, 4]
x3 = [1, 2, 3, 4, 5]
y3 = [7, 2, 1, 4, 5]
plt.plot(x1, y1, label='Line 1', linewidth=2, color='blue')
plt.plot(x2, y2, label='Line 2', linewidth=3, color='red')
plt.plot(x3, y3, label='Line 3', linewidth=1, color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Legends')
plt.legend()
plt.show()
```

```python
# 7) Create plot usingsubplots(),scatter, and some other methods on theAxesobject:
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Scatter plot
axs[0, 0].scatter(x, y, color='blue', alpha=0.5)
axs[0, 0].set_title('Scatter Plot')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# histogram
axs[0, 1].hist(x, bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Histogram')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Frequency')

#bar chart
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 20, 35]
axs[1, 0].bar(categories, values, color='orange', alpha=0.7)
axs[1, 0].set_title('Bar Chart')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Value')

#line chart
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)
axs[1, 1].plot(x_line, y_line, color='red', alpha=0.7)
axs[1, 1].set_title('Line Plot')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

plt.tight_layout()
plt.show()

```

```python
# 9) Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.
products = ["Smartphones", "Laptops", "Tablets", "Smart Home Devices", "Wearables", "Gaming Consoles", "TVs", "Printers"]
sales_units = [5000, 2000, 1500, 3000, 4000, 1000, 500, 200]
plt.figure(figsize=(10, 6))
plt.bar(products, sales_units, color=['blue', 'green', 'orange', 'purple', 'red', 'gold', 'cyan', 'magenta'])
plt.xlabel("Products")
plt.ylabel("Sales Units (Thousands)")
plt.title("Comparison of Product Sales")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
