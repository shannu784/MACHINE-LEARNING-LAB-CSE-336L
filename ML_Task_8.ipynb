### **1) Take two 3D coordinates from the user. Find out the distance between these points using following distance functions. x and y are vectors representing the coordinates of two points.Here, k=3**

```python
import math
#Euclidean distance between the points

x1, y1, z1 = map(int, input("Enter the coordinates of the first point (x y z): ").split())
x2, y2, z2 = map(int, input("Enter the coordinates of the second point (x y z): ").split())

distance = math.sqrt((x2 - x1)*2 + (y2 - y1)*2 + (z2 - z1)*2)

print(f"The distance between the two points is: {distance}")
```

```python
#  Manhattan distance between the points
x1, y1, z1 = map(int, input("Enter the coordinates of the first point (x y z): ").split())
x2, y2, z2 = map(int, input("Enter the coordinates of the second point (x y z): ").split())
distance = abs(x2 - x1) + abs(y2 - y1) + abs(z2 - z1)

print(f"The Manhattan distance between the two points is: {distance}")
```

```python
import math

# Minkowski distance between the points
x1, y1, z1 = map(int, input("Enter the coordinates of the first point (x y z): ").split())
x2, y2, z2 = map(int, input("Enter the coordinates of the second point (x y z): ").split())

p = int(input("Enter the value of p for the Minkowski distance: "))
distance = ((abs(x2 - x1)**p) + (abs(y2 - y1)**p) + (abs(z2 - z1)**p))**(1/p)

print(f"The Minkowski distance between the two points is: {distance}")
```

### **2)Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for all the data samples in the feature space. Distance metric is a 2D array, where the (i,j)th entry represents the distance between the ith and jth sample points in the feature space.**

```python
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances

iris = load_iris()
data = iris.data

distance_matrix = euclidean_distances(data)

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

### **3)Import the Iris dataset. Prepare a dataset considering samples belong to any two output classes. Draw the scatter plot for all the samples in the new dataset considering any two input attributes. Examine the scatter plot to find the equation of a line that can separate sample of two classes.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
new_dataset = pd.concat([class_0, class_1], ignore_index=True)

x_attribute = 'sepal length (cm)'
y_attribute = 'sepal width (cm)'


plt.figure(figsize=(8, 6))
plt.scatter(new_dataset[x_attribute], new_dataset[y_attribute], c=new_dataset['target'], cmap='viridis')
plt.xlabel(x_attribute)
plt.ylabel(y_attribute)
plt.title('Scatter Plot of Samples from Two Output Classes')
plt.colorbar(label='Output Class')
plt.grid(True)
plt.show()

m = -0.5
c = 3
separating_line_equation = f"y = {m}x + {c}"
print("Equation of the separating line:", separating_line_equation)
```
